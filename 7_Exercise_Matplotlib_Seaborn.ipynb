**Q-6** Plot the bargraph between Count of target variable per category.

```python
## Write your code here
ax = sns.countplot(data=df, x='Churn', order=df.Churn.value_counts().index)
#add exact values on top of the bar chart
for i in ax.containers:
    ax.bar_label(i,)
ax.set_xlabel("Churn Type")
ax.set_ylabel("Count of Churn")
ax.set_title("Count of Churn per Category")
```

**Q-14** Build a corelation of all predictors with 'Churn' and plot a bar chart (Use dataframe.corr() function)

```python
## Write your code here
#use pearson correlation
corr=df.corr(method ='pearson')
#take all correlation coeffecient of the predictors only
corr_bar=corr['Churn'].head(3)
#plot bar chart
ax =corr_bar.plot.bar()
#display value on top og the bar
ax.bar_label(ax.containers[0], fmt='{:,.3f}')
ax.set_xlabel("Churn Type")
ax.set_ylabel("Correlation Coeffecient with Churn")
ax.set_title("Distribution of the Churn")
```
